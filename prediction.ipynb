```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import plotly.express as px
import streamlit as st
```

```python
df = pd.read_csv('/Users/um/Desktop/Wild Code School/Protojam/dataOk.csv')
```

```python
df.drop(columns = ['Unnamed: 0'], inplace = True, axis = 1
        )
```

```python
df.head(1)
```

```python
df_france = df[(df['region']== 'France') & (df['Season']== 'Summer')]
```

```python
df_france.info()
```

```python
df = pd.get_dummies(df, columns = ['Medal'], prefix = 'Medal')
```

```python
df = pd.get_dummies(df, columns = ['region'], prefix = 'region')
```

```python
df
```

```python
medal_columns = [col for col in df.columns if 'Medal' in col]
```

```python
region_columns = [col for col in df.columns if 'region' in col]
```

```python
group_df = df.groupby(['Year'])[medal_columns+region_columns].sum().reset_index()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
train_df = group_df[group_df['Year'] < 2020]
test_df = group_df[group_df['Year'] >= 2020]

X_train = train_df.drop(columns=medal_columns)
y_train = train_df[medal_columns]
X_test = test_df.drop(columns=medal_columns)
y_test = test_df[medal_columns]
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, predictions)
mae
```

```python
new_data = pd.DataFrame({
    'Year':[2024]
})

for col in region_columns:
    new_data[col] = 0  # Initialiser à 0
new_data['region_France'] = 1  # Exemple pour la France
```

```python
predicted_medals = model.predict(new_data)
predicted_medals_df = pd.DataFrame(predicted_medals, columns=medal_columns)
predicted_medals_df['Year'] = new_data['Year']
```
